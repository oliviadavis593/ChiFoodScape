```python
import pandas as pd

# Load the spatially joined dataset
df = pd.read_csv("../data/cleaned/grocery_stores_with_community.csv")

# Filter for unclassified stores
unclassified = df[
    (df['IS_REAL_GROCERY'] == False) & 
    (df['IS_JUNK_STORE'] == False)
]

print(f"üïµÔ∏è Unclassified stores: {len(unclassified)}")

# Count unclassified stores by community
unclassified_by_community = unclassified['Community'].value_counts().reset_index()
unclassified_by_community.columns = ['Community', 'UnclassifiedCount']
display(unclassified_by_community.head(10))

# Preview stores in top 3 neighborhoods with unclassified stores
top_areas = unclassified_by_community['Community'].head(3).tolist()

for area in top_areas:
    print(f"\nüß≠ Reviewing: {area}")
    display(
        unclassified[
            unclassified['Community'] == area
        ][['DBA Name', 'Address', 'Zip']]
        .sort_values('DBA Name')
        .head(15)
    )

```
